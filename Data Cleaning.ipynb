# Contract

player's contract details

```python
copy_data['Contract'].dtype
```

```python
copy_data['Contract'].unique()
```

```python
for index , row in copy_data.iterrows():
    if 'On Loan' in row['Contract'] or 'Free' in row['Contract']:
        print(row['Contract'])
```

```python
def extract_contract_info(contract):
    if contract == 'Free' or 'On Loan' in contract:
        start_date = np.nan
        end_date = np.nan
        contract_length = 0
    else:
        start_date, end_date = contract.split(' ~ ')
        start_year = int(start_date[:4])
        end_year = int(end_date[:4])
        contract_length = end_year - start_year
    return start_date, end_date, contract_length

#Apply function to the contract column & create new columns

new_cols = ['Contract Start', 'Contract End', 'Contract Length(years)']
new_data = copy_data['Contract'].apply(lambda x: pd.Series(extract_contract_info(x)))

for i in range(len(new_cols)):
    copy_data.insert(loc=copy_data.columns.get_loc('Contract')+1+i, column=new_cols[i], value=new_data[i])
```

```python
copy_data.sample(5)
```

```python
copy_data[['Contract','Contract Start', 'Contract End', 'Contract Length(years)']].sample(5)
```

```python
#Contract categories

def categorize_contract_status(contract):
    if contract == 'Free' :
        return 'Free'
    elif 'On loan' in contract:
        return 'On Loan'
    else:
        return 'Contract'
    
#Add Contract Status column
copy_data.insert(copy_data.columns.get_loc('Contract Length(years)')+1,'Contract Status', copy_data['Contract'].apply(categorize_contract_status))
copy_data.sample(5)
```

```python
copy_data[['Contract','Contract Start', 'Contract End', 'Contract Length(years)', 'Contract Status']].sample(5)
```

# Height

Player's height in centimeters

```python
copy_data['Height'].dtype
```

```python
copy_data['Height'].unique()
```

```python
def convert_height(height):
    if 'cm' in height:
        return int(height.strip("cm"))
    else:
        feet, inches = height.split("'")
        total_inches = int(feet)*12 + int(inches.strip('"'))
        return round(total_inches * 2.54)
    
#Apply function to height column
copy_data['Height'] = copy_data['Height'].apply(convert_height)
copy_data['Height'].unique()
```

```python
copy_data = copy_data.rename(columns = {'Height':"Heights(cm)"})
copy_data.sample(5)
```

# Weight

Player's weight in kilograms

```python
copy_data['Weight'].dtype
```

```python
copy_data['Weight'].unique()
```

```python
def convert_weight(weight):
    if "kg" in weight:
        return int(weight.strip("kg"))
    else:
        pounds = int(weight.strip("lbs"))
        return round(pounds/2.205)
    
#Apply function to weight column
copy_data['Weight'] = copy_data['Weight'].apply(convert_weight)
copy_data['Weight'].unique()
```

```python
copy_data = copy_data.rename(columns={"Weight":"Weight(kg)"})
copy_data.sample(5)
```
